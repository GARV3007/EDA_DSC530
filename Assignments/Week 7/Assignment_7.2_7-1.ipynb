```python
# Import required libraires
import sys
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
import first
import thinkplot
import thinkstats2
```

```python
# scatter plot of birth weight versus mother’s age. 
live, firsts, others = first.MakeFrames()
ages = live.agepreg
weights = live.totalwgt_lb

plt.scatter(ages, weights, alpha=0.5)
plt.xlabel("age (years)")
plt.ylabel("weight")
```

```python
# Plot percentiles of birth weight versus mother’s age.

# Create dataframe for first babies
live, firsts, others = first.MakeFrames()
live = live.dropna(subset=['agepreg', 'totalwgt_lb'])

# Binning of data
bins = np.arange(10, 48, 3)
indices = np.digitize(live.agepreg, bins)
groups = live.groupby(indices)

# compute the mean height and the CDF of weight for each group
ages = [group.agepreg.mean() for i, group in groups][1:-1]
cdfs = [thinkstats2.Cdf(group.totalwgt_lb) for i, group in groups][1:-1]

# Plot the percentiles of weight versus height
thinkplot.PrePlot(3)
for percent in [75, 50, 25]:
    weights = [cdf.Percentile(percent) for cdf in cdfs]
    label = '%dth' % percent
    thinkplot.Plot(ages, weights, label=label)
thinkplot.Config(xlabel="Mother's age (years)",
                ylabel='Birth weight (lbs)',
                xlim=[14, 45], legend=True)
```

```python
# Compute Pearson’s and Spearman’s correlations
ages = live.agepreg
weights = live.totalwgt_lb
print('thinkstats2 Corr', thinkstats2.Corr(ages, weights))
print('thinkstats2 SpearmanCorr', thinkstats2.SpearmanCorr(ages, weights))
```
