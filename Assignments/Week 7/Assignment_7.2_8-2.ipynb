```python
import numpy as np
import random
import thinkstats2
import thinkplot
```

```python

def SimulateSample(n):
        lam = 2           # value of lambda
        iters = 1000      # number of experiment
        estimates = []    # Dict.

        def VertLine(x, y=1):
                thinkplot.Plot([x, x], [0, y], color='0.8', linewidth=3)

        for _ in range(iters):
            xs = np.random.exponential(1.0/lam, n)       # Generate random sample
            lamhat = 1.0 / np.mean(xs)
            estimates.append(lamhat)

        e2 = [(estimate-lam**2)**2 for estimate in estimates]
        mse2 = np.mean(e2)
        stderr = np.sqrt(mse2)                           # Standard Error

        print('standard error', stderr)

        cdf = thinkstats2.Cdf(estimates)                 # calculate CDF
        ci = cdf.Percentile(5), cdf.Percentile(95)       # Confidence Interval
        print('confidence interval', ci)
        VertLine(ci[0])
        VertLine(ci[1])

        # plot the CDF
        thinkplot.Cdf(cdf)
        thinkplot.Config(xlabel='estimate',
                        ylabel='CDF',
                        title='Sampling distribution')

SimulateSample(10)   #sample size of 10
```

```python
SimulateSample(100)    # sample size of 100
```

```python
SimulateSample(1000)    # sample size of 1000
```

==> **Sample size of n=10 has, standard error = 1.96 and confidence interval (1.234893988308161, 3.6214171308693515) <br>
==> With the increase in sample size, satndard error increases. <br>
==> For all above 3 cases confidence interval size is 2.**
